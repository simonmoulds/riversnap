# Data

This tutorial will show you how to download and organise data so that you can start using riversnap. 

The main purpose of riversnap is to identify the reach (or grid cell) of a river hydrography corresponding to the real-world location of an in situ river gauging station. To start using riversnap, you will minimally require a river hydrography dataset and a set of points showing the location of gauging stations. 

## Hydrography data

Multiple river hydrography datasets exist. Currently, riversnap supports the Global River Topology (GRIT), RiverAtlas, SWORD, MERIT-Hydro, and TDX-Hydro. The subsequent examples will use GRIT, which can be downloaded from [Zenodo](https://doi.org/10.5281/zenodo.7629907). 

For this example we will retrieve [GRITv1.0_segments_EU_EPSG4326.gpkg.zip](https://zenodo.org/records/17435232/files/GRITv1.0_segments_EU_EPSG4326.gpkg.zip?download=1). This archive contains a GeoPackage file (`.gpkg`) with river segments for Europe in the EPSG:4326 coordinate reference system. 

Save the file in a sensible location (e.g. a directory called `GRIT`). After extracting the `.zip`, you should end up with something like this: 

```
GRIT/ 
    GRITv1.0_segments_EU_EPSG4326.gpkg.zip
    GRITv1.0_segments_EU_EPSG4326.gpkg
```

## Gauge data

Where riversnap is strict about the supported river hydrography datasets, it is more flexible regarding the in situ river gauging stations. In the simplest case all that is needed is a georeferenced set of points with unique IDs, although more sophisticated snapping routines can make use of attribute data (e.g. mean river discharge, catchment area). 

In the examples we will use the EStreams dataset, available from [Zenodo](https://doi.org/10.5281/zenodo.10733141). Download the [shapefiles.zip](https://zenodo.org/records/17598150/files/shapefiles.zip?download=1) archive. 

Save the file somewhere sensible, then extract the `.zip`. You should end up with a shapefile dataset like this:

```
EStreams/
    shapefiles.zip
    shapefiles/ 
        estreams_gauging_stations.cpg
        estreams_gauging_stations.dbf
        estreams_gauging_stations.prj
        estreams_gauging_stations.shp
        estreams_gauging_stations.shx
```
